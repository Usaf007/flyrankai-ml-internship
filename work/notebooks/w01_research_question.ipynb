## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

# Loading the starter dataset directly from the raw GitHub URL so Colab can read it
url = "https://raw.githubusercontent.com/flyrank-bih/flyrank-ml-internship-starter/main/data/raw/content_refresh_anonymized.csv"
df = pd.read_csv(url)

# Calculating the base rate of the proxy label
decay_rate = df["trend_direction"].str.lower().eq("down").mean()

# Finding the median age of pages in the dataset
median_age = df["content_age_days"].median()

print("--- DATA JUSTIFICATION FOR THIS LANE ---")
print(f"1. The Problem Scale: {decay_rate*100:.1f}% of the tracked pages in this dataset are currently in a downward trend. A ranked queue is mandatory to sift through this volume.")
print(f"2. The Age Factor: The median content age in this inventory is {median_age:.0f} days. With an aging content library, identifying which specific stale pages are actually losing traffic is a critical decision-support need.")
```
