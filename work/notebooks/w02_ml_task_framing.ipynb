## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule (like `content_age > 180 days` AND `impressions > 500`) is brittle. It creates massive blocks of tied scores and ignores nuance.

Machine Learning beats a fixed rule here because a model (like a Random Forest) can weigh dozens of continuous variables simultaneously—such as the exact gap in CTR, precise age, and engagement rate—to find subtle decay patterns that rigid human logic misses. It provides a granular, probabilistic score rather than a simple Yes/No, allowing for much better prioritization.
